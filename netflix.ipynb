```python
# Importando as bibliotecas que serão utilizadas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leitura do arquivo csv

%time netflix = pd.read_csv('netflix_titles_2021.csv')
%time netflix.head(10)
```

```python
#verificando o tamanho do dataset

%time netflix.shape
```

```python
# Alterando os nomes das colunas

colunas = {'type':'tipo', 'title':'titulo', 'director':'diretor', 'cast':'elenco',
            'country':'pais', 'date_added':'adicionado_em','release_year':'ano_lancamento',
            'rating':'avaliacao', 'duration':'duracao', 'listed_in':'categorias', 'description':'descricao'}

%time netflix.rename(columns=colunas, inplace=True)
```

```python
%time netflix.head()
```

```python
# Contagem/% de valores nulos no dataframe

%time frequencia = netflix.isna().sum() 
%time percentual_na = round((netflix.isna().sum() / len(netflix))*100, 2)
%time valores_na = pd.DataFrame([frequencia, percentual_na]).transpose().rename(columns={0:'Frequência', 1:'% de NAs'})
valores_na
```

```python
%time netflix.info()
```

```python
# Otimizando o tamanho da variável ano_lancamento

%time netflix['ano_lancamento'] = netflix["ano_lancamento"].astype('int16')
```

```python
netflix.info()
```

```python
# Cria um dicionário com os valores dos meses

import calendar

%time meses = dict((month, str(index)) for index, month in enumerate(calendar.month_name) if month)
meses
```

```python
# modificando o texto para achar um padrão para o split 
%time netflix['adicionado_em'] = netflix.adicionado_em.str.strip()
%time netflix['adicionado_em'] = netflix['adicionado_em'].str.replace(', ', '/')
%time netflix['adicionado_em'] = netflix['adicionado_em'].str.replace(' ', '/')
```

```python
# separação do valor adicionado_em em dia, mês e ano

%time data = netflix['adicionado_em'].str.split('/', n = 2, expand = True)
%time mes = data[0]
%time dia = data[1]
%time ano = data[2]
```

```python
# substituição dos nomes dos meses por seus respectivos valores

%time mes.replace(meses, inplace=True)
```

```python
# novo dataframe pandas para inclusão da coluna de data

%time data = pd.DataFrame([dia, mes, ano]).transpose().rename(columns={1:'dia', 0:'mes', 2:'ano'})
%time data.dia = data.dia.str.zfill(2)
%time data.mes = data.mes.str.zfill(2)
%time data['data'] = data.dia + '/' + data.mes + '/' + data.ano
data
```

```python
# remoção das colunas dia, mes e ano
%time data.drop(columns=['dia', 'mes', 'ano'], inplace=True)
data
```

```python
# modifica a coluna adicionado_em para valores com datas no formato dd/mm/yyyy.
%time netflix['adicionado_em'] = data
netflix.head(5)
```

```python
%time netflix['adicionado_em'] = netflix['adicionado_em'].astype('datetime64[ns]')
```

```python
netflix.info()
```

```python
# Agrupando os valores por categorias

%time categorias = netflix['categorias'].value_counts()
categorias
```

```python
# Top 3 categorias

%time top_3_categorias = categorias.head(3)
top_3_categorias
```

```python
%%time

# função para add valores nas barras 
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2, y[i], ha = 'center', color = 'white')


plt.bar(x=top_3_categorias.index, height=top_3_categorias)
plt.xlabel('categorias')
plt.title('Top 3 Categorias')
addlabels(x=top_3_categorias.index, y=top_3_categorias)
plt.show()
```

```python
# Top 5 Países
paises = netflix.pais.value_counts().head(5)
paises
label = paises.index
myexplode = [0.2, 0, 0, 0, 0]
plt.pie(paises, labels=label, explode=myexplode)
plt.title('Top 5 países')
plt.show()
```

```python
# retirnado os valores nulos do dataset

%time netflix_transformado = netflix.dropna()
%time netflix_transformado.head(5)
```

```python
%time netflix_transformado.info()
```

```python
# separando o dataset por duração (Season)

%time selecao = netflix_transformado.duracao.str.contains('Season')
%time season = netflix_transformado[selecao]
season.head()
```

```python
season.info()
```

```python
# separando o dataset por duração (Minutos)

%time selecao = netflix_transformado.duracao.str.contains('min')
%time filmes_minutos = netflix_transformado[selecao]
filmes_minutos.reset_index(inplace=True, drop=True)
filmes_minutos.head()
```

```python
%time filmes_minutos.info()
```

```python
# tranformando o valor da duração em minutos (str) para int
%time filmes_minutos['duracao'] = filmes_minutos.loc[:, 'duracao'].str.replace(' min', '')
filmes_minutos.head(5)
```

```python
%time filmes_minutos.duracao = filmes_minutos.loc[:, 'duracao'].astype('int16')
```

```python
filmes_minutos.info()
```

```python
# Média de duração dos filmes (min)

%time media_duracao = round(filmes_minutos.duracao.mean(), 2)
media_duracao
```

```python
# Verificando as estatísticas básicas do dataset filmes_minutos

%time filmes_minutos.duracao.describe()
```

```python
%%time

# boxplot - verificando outliers

bp = plt.boxplot(x='duracao', data=filmes_minutos, vert=1, patch_artist=True, labels=['Duração'])
colors = 'blue'
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
for median in bp['medians']:
    median.set(color ='red',
               linewidth = 3)
for whisker in bp['whiskers']:
    whisker.set(color ='red',
                linewidth = 1.5,
                linestyle =":")
plt.show()
```
